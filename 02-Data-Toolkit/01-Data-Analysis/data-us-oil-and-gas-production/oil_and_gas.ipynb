## Yearly Gas production

```python
# How much gas has been produced yearly by each US state, and by the US as a whole?

gas_df_numeric = gas_df.select_dtypes(include=np.number)
yearly_gas_df = gas_df_numeric.groupby(gas_df['Month'].dt.year).sum()
```

### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]

from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

```python
plot = yearly_gas_df.filter(items=["U.S."]).plot(kind="bar")
plot.set_xlabel("Year");
```

```python
filtered_yearly_gas_df = yearly_gas_df.loc[2009:2017]
filtered_yearly_gas_df
```

## Comparing with Crude Oil Production

```python
oil_df = pd.read_csv("data/U.S._crude_oil_production.csv", parse_dates=["Month"])
oil_df.head()
```

---

## Yearly Oil production

```python
oil_df_numeric = oil_df.select_dtypes(include=np.number)
yearly_oil_df = oil_df_numeric.groupby(oil_df['Month'].dt.year).sum()
yearly_oil_df.head()
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.rstrip()
```

```python
yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar');
```

```python
filtered_yearly_oil_df = yearly_oil_df.loc[2009:2017]
filtered_yearly_oil_df
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```

## Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df[["U.S."]].rename(columns={"U.S.": "Gas"})
total_oil = filtered_yearly_oil_df[["U.S. Crude Oil"]].rename(columns={"U.S. Crude Oil": "Crude Oil"})
```

```python
total_gas
```

```python
merged_df = pd.concat([total_gas, total_oil], axis=1)
```

```python
merged_df
```

```python
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
